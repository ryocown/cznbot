```
from ultralytics import YOLO

# 1. Load the fastest base model
model = YOLO('yolov8n.pt')
```

```
# 2. Train the model using your exported dataset
# Replace 'data.yaml' with the path to your dataset config.
results = model.train(data='../data/annotated/data.yaml', epochs=50, imgsz=640)
```

```
# 3. Export to TensorFlow Lite for Android
model.export(format='tflite')
```
